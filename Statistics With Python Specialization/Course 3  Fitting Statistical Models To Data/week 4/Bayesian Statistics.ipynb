We have primarily used Maximum likelihood to fit models ("Pick the ***best*** values to fit the data")

Probability, in a frequentist setting is a long run frequency, the Long run ecpection in "If i do this many times."

Bayesian methods think about these problems in a fundamentally different way.
![hmm](week-4-img/to-bayesian.png)

![hmm](week-4-img/belief-n-bayesian.png)
This is how bayesians think of the world.

This new distribution is the posterior.
![hmm](week-4-img/posterior.png)

If we were doing the same in a frequentist analysis, we would get a point estimate:


![hmm](week-4-img/freq-vs-bayes.png)

**In a frequentist setting: we would get 1 point, say 118. In a bayesian scenario, we would get an entire distribution. That distribution is called a posterior. The posterior allows us to update our belief and answer questions about our quantity of interest.**

![hmm](week-4-img/the-posterior.png)

So with the posterior we can ask questions and say what we believe about the mean, etc.

![hmm](week-4-img/posterior-questions.png)

* In frequentist setting, we can say:
    * With 95% confidence we can say that, if we were to carry out this experiment many times, the true mean IQ of UoM students falls between the values (116.64,122.26).
* In a bayesian belief, we can say: (talks of probability instead of confidence)
    * There is a 95% probability that the belief is between (116.64,122.26).
    * We can say different things becasue we think of them in different ways
    * We get all of that from the posterior distributation 

**Being able to talk about probablity is very powerful. But nothing comes for free. The Bayesian statistics has issues as well**

![hmm](week-4-img/bayesian-working.png)


![hmm](week-4-img/bayesian-cycle.png)

![hmm](week-4-img/bayesian-steps.png)


### But why?
Because all the values we want to know, all questions about our beliefs or quantity of interest can be found via the posterior. <br>
We can combine this with a loss function for optimal decision making and we can really use this to get a better model of the world. <br>
This is a very powerful idea. Because we now have distributions and not just point estimates, we can combine this with ways of optimizing loss for better decision making under uncertainty.
<br>
**But this method has some downsides** <br>
1. We have to rewrite the entire definition of probability to make this work
2. In mathematics, this problem can be very difficult.  It can sometimes even be intractable, and at the very least , is usually pretty computationally expensive depending on the model.
