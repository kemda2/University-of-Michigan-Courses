## Linear Regression Inference

We used the least squared method to create our regression line.

**Regression Inference: Is tehre a significant (positive) linear relationship between CW Distance and Height?**

* If our slope is 0, no matter the values of our independent variables (x), we will predict the exact same value for our dependent variable(y).
    * So, Knowing the value of x is not really helping us to predict y.

* Our slope b1 = 1.1 is only an estimated slope.
    * This means if we were to carry out the same exact study with different adults, we will not get this same exact estimate.

![regression inference ](week-2-img/linear-regression-inference-1.png)

#### Checking the values from the model:
b1 (coeff) = 1.1076 <br>

It's not = 0 but we are interested in seeing how far away from 0 is it<br>

Std-err = 0.670 <br>

This shows how far away the estimated slopes like the one we got are going to be away from the true slope on average<br>

t-statistic = 1.653 <br>

taking into account the above two information, we form our t-statistic. It is measuring how close is our estimated slope of 1.1 from 0 in standard error units. So our estimated slope 1.1 was 1.65 standard error units above 0<br>

P>|t| = 0.112 <br>

Converting the test statistic to a P value (probability value), we get .112 which is not that small, but this is the p-value for a 2 sided alternative. Since we are only interested in knowing whether the true slope is greater than zero or not, we will divide it by 2 getting 0.056. Since our research question was a significant positive rel bw 2 vars, our alternative would be >0. It is significant at a 10% level but not at a 5% level marginally<br>

**Underlying assumptions:**
* We fit a regression model, we fit CW Distance on height
* Individual CW Distance are listed as being a linear function of height (true mean response at a given height)
* individual CW Dist will vary from the mean and so we have the error, e the random error
    It should be normally distributed around zero and should have a constant variance e ~ N(0, sigma^2)
![reg inf 1](week-2-img/underlying-assumptions.png)

**The true errors need to be normally distributed. But! We don't have a true regression line or the true errors. We have an estimated regression line from data, the observed terms (the residuals i.e realized values of e) which we can then put into graphs and check for normality.**
